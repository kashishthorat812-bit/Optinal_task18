```python
import matplotlib.pyplot as plt
sns.histplot(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show
```

```python
import matplotlib.pyplot as plt
sns.histplot(df["bmi"])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show
```

```python
import matplotlib.pyplot as plt
sns.histplot(df["children"])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Frequency")
plt.show
```

```python
import matplotlib.pyplot as plt
sns.histplot(df["expenses"])
plt.title("Expenses Distribution")
plt.xlabel("Expenses")
plt.ylabel("Frequency")
plt.show
```

```python
import matplotlib.pyplot as plt
sns.countplot(x="sex",data=df)
plt.title("Sex count")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show
```

```python
import matplotlib.pyplot as plt
sns.countplot(x="smoker",data=df)
plt.title("Smoker count")
plt.xlabel("smoker")
plt.ylabel("Count")
plt.show
```

```python
import matplotlib.pyplot as plt
sns.countplot(x="region",data=df)
plt.title("Region count")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

```python
# insurance Expenses
print("Average Insurance Expenses :")
print(df["expenses"].mean())

# print the Maximum Expenses
print("Maximum Expenses :")
print(df["expenses"].max())

# print the Minimum Expenses
print("Minimum Expenses :")
print(df["expenses"].min())

# average Expenses for smoker vr non smoker
print("\nAverage Expenses for smoker vr non smoker :")
print(df.groupby("smoker")["expenses"].mean())
```

```python
sns.boxplot(x=df["smoker"],y=df["expenses"])
plt.title("Expenses by smoker")
plt.xlabel("smoker")
plt.ylabel("expenses")
plt.show
```

```python
sns.boxplot(x=df["sex"],y=df["expenses"])
plt.title("Expenses by sex")
plt.xlabel("sex")
plt.ylabel("expenses")
plt.show
```
